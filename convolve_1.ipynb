```python
X_train_part,X_val,y_train_part,y_val = train_test_split(X_train_scaled,y_train,random_state = 42,test_size = 0.42)
```

```python
def objective(trial):
    # 1. TUNING PARAMETERS (Optimized for Speed & RAM Safety)
    params = {
        # Keep trees FEW during tuning (e.g., 20-50).
        # You increase this to 500+ ONLY for the final training.
        'n_estimators': trial.suggest_int('n_estimators', 60, 200),

        # CRITICAL: Depth must be limited.
        # Depth > 20 on 2M rows explodes memory usage.
        'max_depth': trial.suggest_int('max_depth', 15, 50),

        # CRITICAL: Do NOT let trees see all 2M rows.
        # '0.1' means each tree trains on 10% of data (200k rows).
        # This is the secret to preventing crashes.
        'max_samples': trial.suggest_float('max_samples', 0.3, 0.5),

        # Standard tuning
        'min_samples_split': trial.suggest_int('min_samples_split', 5, 30),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 2, 20),
        'max_features': trial.suggest_float('max_features', 0.3, 0.7),

        # Mandatory settings for large data
        'bootstrap': True,
        'n_jobs': 4,        # Limit cores to 4 to save RAM (don't use -1 here)
        'random_state': 42
    }

    # 2. CREATE MODEL
    model = RandomForestRegressor(**params)

    # 3. TRAIN & EVALUATE
    # We train on the partial set (X_train_part) to be fast
    model.fit(X_train_part, y_train_part)

    preds = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, preds))

    return rmse
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=20)
```

```python
# study = optuna.create_study(direction='minimize')
# study.optimize(objective, n_trials =10)
```

```python
best_params = study.best_params
final_model = RandomForestRegressor(**best_params, n_jobs=-1, random_state=42)
final_model.fit(X_train_scaled,y_train)
y_pred = final_model.predict(X_test_sc)
```
